# Electric Fields Lab

## Objective: 
Experimentally determine the equation for the electric field between two concentric circles with center conductor $ v=\theta$ volts

Assume relation is of form 

$ E = \dfrac{1}{r^n}$

($n=1$ Accepted)

r = radius

k = constant

n = exponent. <-- **What we want to find**

Put E into linear form:

$ ln(\dfrac{E}{E_0}) = -n(ln(\dfrac{r}{r_0})) + ln(\dfrac{E^*}{E_0})$

$             y       =       m       x       +            b $

where:

$E_0 = 1 \frac{V}{m} $

$r_0^n = 1 m$

$E^* = \dfrac{k}{r_0^n}$


This makes the equation dimensionless so you can take ln. Nothing to calculate here!


# Procedure
- Set power supply to ~12 Volts(Check with meter)

$ E(r) = \dfrac{\delta V}{\delta r}$

```python
def rulefour(Q, const, delt, vals):
    return np.abs(Q)*np.sqrt(np.sum((const*delt/vals)**2))
    
```
